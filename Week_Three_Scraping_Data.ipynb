```python
import requests
from bs4 import BeautifulSoup
import csv
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
try:
    response = requests.get("https://books.toscrape.com/")
    soup = BeautifulSoup(response.content, 'html.parser')
    books = soup.find_all("article", class_ = 'product_pod')

    data = []
    for book in books[:20]:
        title = book.h3.a['title']
        price = book.find('p', class_ = 'price_color').text

        data.append([title, price])


    api_key = "b7ec9c1763dda5d2b4c6c309"
    original_currency = input("Enter the original currency. (Please note that the default currency is in pounds(GBP)): ").upper()
    expected_currency = input("Enter the expected currency: ").upper()

    url = f"https://v6.exchangerate-api.com/v6/{api_key}/latest/{original_currency}"
    response = requests.get(url)
    rates = response.json()['conversion_rates']


    for i in range(len(data)):
        gbp_price = float(data[i][1].replace('Â£', ''))
        kes_price = gbp_price * rates[expected_currency]
        data[i][1] = round(kes_price, 2)

except:
    print("Error fetching results")

data
```

```python
with open('Book_store_Prices_in_Ksh.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Price"])
    writer.writerows(data)

print("Data has been written to Book_store_prices_in_ksh.csv successfully")
```
